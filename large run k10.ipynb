Simulation with $N=576$
------------------------
This is a simulation with a larger grid, namely $N=576$ for every dimension. <br>
Slices are shown below.

```python
import matplotlib
%matplotlib inline
```

```python
matplotlib.rcParams.update({'font.size': 22, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
import pencil as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1, magpow = pc.read_power('power_mag.dat')
t2, kinpow = pc.read_power('power_kin.dat')
```

```python
pars = pc.read_param(quiet=True)
```

```python
par2 = pc.read_param(quiet=True, param2=True)
dim = pc.read_dim()
print par2.force, dim.nxgrid
```

```python
kmodes = range(1,dim.nxgrid/2+1)

fig = plt.figure(figsize=(10,12))

ax1 = fig.add_subplot(211)
ax1.set_title('Spectra at t={0:.0f}'.format(t1[0]))
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylim(1.0e-12, 0.1)
ax1.set_xlim(1,dim.nxgrid/2)
ax1.set_title('Spectra at t={0:.0f}'.format(t1[0]))

ax1.plot(magpow[0,:], label='$E_{mag}$')
ax1.plot(kinpow[0,:], label='$E_{kin}$')
ax1.plot(range(1,10), [1.0e-5*i**2 for i in range(1,10)], color='red')
ax1.plot(range(1,10), [1.0e-7*i**4 for i in range(1,10)], color='purple')
ax1.text(2,2e-4,'$\propto\ k^{2}$', color='red')
ax1.text(3,2e-6,'$\propto\ k^{4}$', color='purple')
ax1.legend(loc=1,prop={'size':18})

ax2 = fig.add_subplot(212)
ax2.set_title('Spectra at t={0:.0f}'.format(t1[-1]))
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylim(1.0e-10, 0.1)
ax2.set_xlim(1,dim.nxgrid/2)

ax2.plot(magpow[-1,:], label='$E_{mag}$')
ax2.plot(kinpow[-1,:], label='$E_{kin}$')
ax2.plot(range(1,10), [3.0e-5*i**(3./2.) for i in range(1,10)])
ax2.text(2,2.5e-5,'$\propto\ k^{2}$', color='red')
ax2.legend(loc=1, prop={'size':18})

#fig.suptitle('High resolution' )
```

```python
vid_uu, t = pc.read_slices(field='uu1')
vid_bb, t = pc.read_slices(field='bb1')
```

```python
print t, vid_uu.shape
```

```python
fig4 = plt.figure(figsize=(12,6))
ext = [-np.pi,np.pi, -np.pi, np.pi]
ax4 = fig4.add_axes([0.1,0.1,0.5,0.9])
ax5 = fig4.add_axes([0.6,0.1,0.9,0.9])

ax4.imshow(vid_uu[-1,:,:], extent=ext, cmap=plt.cm.jet)
ax4.set_xlabel('x [code units]')
ax4.set_ylabel('y [code units]')
ax4.set_title('Velocity Field $u_x$')

ax5.imshow(vid_bb[-1,:,:], extent=ext, cmap=plt.cm.hot)
ax5.set_title('Magnetic Field $B_x$')
ax5.set_xlabel('x [code units]')
ax5.set_ylabel('y [code units]')
```
