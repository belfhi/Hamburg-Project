```python
def adjust_specaxes(ax, ddir='data/'):
    ax.set_xscale('log')
    ax.set_yscale('log')
    dim = pc.read_dim(datadir=ddir)
    ax.set_xlim(1, dim.nx/2)
    ax.set_xlabel('k mode')
    return ax
```

```python
fig = plt.figure(figsize=(12,7), dpi=100)
ax1 = fig.add_subplot(111)
ax1 = adjust_specaxes(ax1)
ax1.set_ylim(1e-8,1e-2)
ax1.set_ylabel('$E_{{kin}}\\ & E_{{mag}}$')
for b in range(powerb.shape[0]):
    if tb[b] <= 1 and b%2 ==0 or tb[b]> 1 and b%10 == 0:
        ax1.plot(powerb[b,:],"-", label='$E_{{mag}},\ t=%.1f$' %tb[b], linewidth=2)
#        ax1.plot(poweru[b,:],"--",  label='$E_{{kin}},\ t=%.0f$' %tu[b], linewidth=3)
ax1.plot(range(1,25), [2e-9*x**4 for x in range(1,25)], color='black', linewidth=3)  
ax1.text(3.5,1e-7, '$B_0\propto k^4$')
ax1.legend(loc='center left', bbox_to_anchor=(1.,.5), prop={'size':16}, frameon=False, ncol=2)
fig.suptitle('Magnetic Spectra', fontsize=24)
fig.savefig('ub_equiv_magnetic.pdf', bbox_inches='tight')
```

```python
tser = pc.read_ts(plot_data=False)
```

```python
values = np.zeros(shape=(powerb.shape[0]-1,5))
for i,pb in enumerate(powerb[:-1,:]):
    maxk = np.where(pb == pb.max())[0][0]
    try:
        eps = 0.01
        tind = np.where((tser.t+eps >tb[i]) & (tser.t-eps < tb[i]))[0][0]
    except IndexError:
        eps = 0.03
        tind = np.where((tser.t+eps >tb[i]) & (tser.t-eps < tb[i]))[0][0]
    taueddy = (1./maxk)/tser.vArms[tind]
    values[i,:] = tser.t[tind], tind, tser.vArms[tind], maxk, taueddy
```

```python
def find_index(time, tarray):
    if time > 5.0:
        rr = 0
    else:
        rr = 1
    try:
        return np.where(np.round(tarray,rr) == round(time,rr))[0][0]
    except IndexError:
        return -1
```

```python
mfig = plt.figure(figsize=(12,7))
maxes = mfig.add_subplot(111)
maxes.plot(values[:,0], values[:,-1])
maxes.set_yscale('log')
maxes.set_xlim(0,100)
maxes.set_ylim(6e-2, 2e1)
#print([round(v,2) for v in values[:100,-1]])
mfig.suptitle('Eddy turnover time', fontsize=24)
```

```python
dspec = []
for tt in range(len(tb)-1):
    dspec.append(round(tb[tt+1]-tb[tt],1))
time = 0.
plottimes = [0.]
i = 0
ii = [0]
while True:
    #print(time)
    time += round(values[i,-1],1)
    plottimes.append(time)
    i = find_index(time, tb)
    ii.append(i)
    if time > tb[-1]:
        break
```

```python
tspecfig = plt.figure(figsize=(12,7))
spax = tspecfig.add_subplot(111)
spax = adjust_specaxes(spax)
spax.set_ylim(1e-9, 1e-2)
for j,i in enumerate(ii):
    spax.plot(powerb[i,:], label='$%d\\cdot\\tau_{{eddy}}$' %j)
spax.legend(loc='center left', bbox_to_anchor=(1.,.5), frameon=False, ncol=3, prop={'size':14})
tspecfig.suptitle('Magnetic Spectra, plotted each Eddy turnover time', fontsize=24)
tspecfig.savefig('equiv_taueddy.png', bbox_inches='tight')
```

```python
def fitfunc(x,a,b):
    return a*x**b
```

```python
tsfig = plt.figure(figsize=(10,7), dpi=100)
tax1 = tsfig.add_subplot(111)
tax1.set_xscale('log')
tax1.set_yscale('log')
tax1.set_xlim(1e-2,100)
tax1.set_ylim(8e-3, 1)
tax1.plot(tser.t, tser.brms, label='$B_{rms}$')
tax1.plot(tser.t, tser.urms, label='$u_{rms}$')
tax1.plot(tser.t[1:], fitfunc(tser.t[1:], 0.09, -0.5))
tax1.plot(tser.t[1:], fitfunc(tser.t[1:], 0.18,  -0.5))
tax1.text(10, 6e-2, '$B_{rms}\propto t^{-1/2}$', color='blue')
tax1.text(0.4, 4e-2, '$u_{rms}\propto t^{-1/2}$', color='green')
#tax1.plot(tser.t, tser.ab_int, label='$\mathcal{H}_{int}$')
tax1.legend(loc='lower center', ncol=3, frameon=False)
tsfig.suptitle('Time Series, Decay Laws', fontsize=24)
```
