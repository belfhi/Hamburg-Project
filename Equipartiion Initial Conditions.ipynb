```python
import matplotlib as mpl
mpl.rcParams.update({'font.size': 24, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
tu, poweru = pc.read_power('poweru.dat')
tb, powerb = pc.read_power('powerb.dat')
```

```python
fig = plt.figure(figsize=(12,7), dpi=100)
ax1 = fig.add_subplot(111)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(1,120)
ax1.set_ylim(1e-8,1e-2)
ax1.set_xlabel('k mode')
ax1.set_ylabel('$E_{{kin}}\\ & E_{{mag}}$')
for b in range(powerb.shape[0]):
    if tb[b] <= 1 and b%2 ==0 or tb[b]> 1 and b%10 == 0:
        ax1.plot(powerb[b,:],"-", label='$E_{{mag}},\ t=%.1f$' %tb[b], linewidth=2)
#        ax1.plot(poweru[b,:],"--",  label='$E_{{kin}},\ t=%.0f$' %tu[b], linewidth=3)
ax1.plot(range(1,25), [2e-9*x**4 for x in range(1,25)], color='black', linewidth=3)  
ax1.text(3.5,1e-7, '$B_0\propto k^4$')
ax1.legend(loc='center left', bbox_to_anchor=(1.,.5), prop={'size':16}, frameon=False, ncol=2)
fig.suptitle('Magnetic Spectra', fontsize=24)
fig.savefig('ub_equiv_magnetic.pdf', bbox_inches='tight')
```

```python
fig2 = plt.figure(figsize=(12,7), dpi=100)
ax2 = fig2.add_subplot(111)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(1,120)
ax2.set_ylim(1e-8,1e-2)
ax2.set_xlabel('k mode')
ax2.set_ylabel('$E_{{kin}}\\ & E_{{mag}}$')
for u,pu in enumerate(poweru):
    if u%10 == 0:
#        ax1.plot(powerb[b,:],"-", label='$E_{{mag}},\ t=%.0f$' %tb[b], linewidth=2)
        ax2.plot(pu,  label='$E_{{kin}},\ t=%.0f$' %tu[u], linewidth=2)
ax2.plot(range(1,25), [3e-6*x**2 for x in range(1,25)], color='black', linewidth=3)   
ax2.text(2,3e-5, '$u_0 \propto k^2$')
ax2.legend(loc='center left', bbox_to_anchor=(1.,.5), prop={'size':16}, frameon=False, ncol=1)
fig2.suptitle('Kinetic Spectra', fontsize=24)
fig2.savefig('ub_equiv_kinetic.pdf', bbox_inches='tight')
```

```python
fig3 = plt.figure(figsize=(12,7), dpi=100)
ax3 = fig3.add_subplot(111)
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_xlim(1,120)
ax3.set_ylim(1e-8,1e-2)
ax3.set_xlabel('k mode')
ax3.set_ylabel('$E_{{kin}}\\ & E_{{mag}}$')
clr=iter(plt.cm.YlGnBu_r(np.linspace(0,1,14)))
clr2=iter(plt.cm.hot(np.linspace(0,1,17)))
for u in range(poweru.shape[0]):
    if u%5 == 0 and tu[u] < 5.9:
        c = next(clr)
        c2 = next(clr2)
        ax3.plot(powerb[u,:],"-", label='$E_{{mag}},\ t=%.1f$' %tb[u], linewidth=2, color=c)
        ax3.plot(poweru[u,:],'-', label='$E_{{kin}},\ t=%.1f$' %tu[u], linewidth=2, color=c2)

ax3.plot(range(1,25), [3e-6*x**2 for x in range(1,25)], color='black', linewidth=3)      
ax3.plot(range(1,25), [2e-9*x**4 for x in range(1,25)], color='black', linewidth=3)  
ax3.text(3.5,1e-7, '$B_0\propto k^4$')
ax3.text(2,3e-5, '$u_0 \propto k^2$')
#ax3.legend(loc='center left', bbox_to_anchor=(1.,.5), prop={'size':16},ncol=2)
ax3.text(10,1.5e-8, '$0 \leq t \leq 5, \ \Delta t=0.5$')
fig3.suptitle('Magnetic & Kinetic Spectra, early decay', fontsize=24)
fig3.savefig('ub_equiv.png', bbox_inches='tight')
```

```python
tser = pc.read_ts(plot_data=False)
```

```python
def fitfunc(x,a,b):
    return a*x**b
```

```python
tsfig = plt.figure(figsize=(10,7), dpi=100)
tax1 = tsfig.add_subplot(111)
tax1.set_xscale('log')
tax1.set_yscale('log')
tax1.set_xlim(1e-2,100)
tax1.set_ylim(8e-3, 1)
tax1.plot(tser.t, tser.brms, label='$B_{rms}$')
tax1.plot(tser.t, tser.urms, label='$u_{rms}$')
tax1.plot(tser.t[1:], fitfunc(tser.t[1:], 0.09, -0.5))
tax1.plot(tser.t[1:], fitfunc(tser.t[1:], 0.18,  -0.5))
tax1.text(10, 6e-2, '$B_{rms}\propto t^{-1/2}$', color='blue')
tax1.text(0.4, 4e-2, '$u_{rms}\propto t^{-1/2}$', color='green')
#tax1.plot(tser.t, tser.ab_int, label='$\mathcal{H}_{int}$')
tax1.legend(loc='lower center', ncol=3, frameon=False)
tsfig.suptitle('Time Series, Decay Laws', fontsize=24)
```
