Comparison of different forcing
------------------------------
In this notebook all the different simulation with different forcing are compared.
The forcing is applied for different wavenumbers $k$. The smallest possible wavenumber is
$k_0=2\pi/L$. In this case the box length $L$ is chosen conveniently: $L=2\pi$ such that $k=1,2,...,N/2$. 
Where $N$ is the number of grid points in the direction. 

At first some Python / Matplotlib imports have to happen so the plotting can begin. <br>
Then the run directories are read and sorted for usage later

```python
import matplotlib
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import pencil as pc
import numpy as np
import os
```

```python
dirlist = filter(lambda names: names.startswith('rerun_k'), os.listdir('.'))
dirlist = sorted(dirlist, key=lambda k: float(k.split('rerun_k')[1]))
print dirlist
```

```python
matplotlib.rcParams.update({'font.size': 22, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

The first Plot is the Comparison of Kinetic energy Spectra for different forcing $k_{av}$
--------------------------------------------------------------------------------------------

<h7>
For this the data has to be read from every run's data directory and then be plotted into the graph.<br>
Note that there is excess energy in the region where the forcing is active.
</h7>

```python
fig1 = plt.figure(figsize=(12,8), dpi=100)
ax1 = fig1.add_subplot(111)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylim(1.0e-10, 1.0e-5)
ax1.set_xlim(1,60)
eps = 0.5
force = 1.0e-3
for dir in dirlist:
    ddir = os.path.join(dir, 'data')
    #pars2 = pc.read_param(datadir=ddir,quiet=True, param2=True)
    #print pars2.tforce_stop
    #if (pars2.force != force):
    #    print pars2.force
    t, power = pc.read_power('power_kin.dat', datadir=ddir)
    pind = np.where( (t < 50+eps) & (t > 50-eps) )[0][0]
    #print pind
    ax1.plot(power[pind,:], label='k={0},  t={1:.0f}'.format(dir.split('_')[1][1:], t[pind]))
    
#ax1.text()
ax1.legend(loc=3, prop={'size':16})
ax1.set_title('Comparison of kinetic energy spectra')
ax1.set_xlabel('k modes')
ax1.set_ylabel('$E_{kin}$')
```

Magnetic energy Spectra
---------------------------------

The same plot as for the kinetic energy is produced. <br>
As the forcing happens in the equations of momentum for the magnetic potential $\textbf{A}$ there is a large peak at the forcing wavenumber $k_{av}$ <br>
The peaks also seem to have a varying width. This is not the case as the x-Axis is logarithmic and we only have energy at full integer values of $k$. <br>
Here two different times are shown, as they have a slightly different behaviour for $1 < k < 30$. <br>
At early times (Here shown for t=5) The slope rises with $E \propto\ k^3$ whereas later it only has a $E\propto\ k^2$ slope.

```python
#setting up the figure
fig2 = plt.figure(figsize=(10,12), dpi=100)
fig2.suptitle('Comparison of Different Forcing values $k_{av}$', fontsize=28)
ax3 = fig2.add_subplot(211)
ax2 = fig2.add_subplot(212)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylim(1.0e-12, 1.0e-3)
ax2.set_xlim(1.,60)
ax2.set_xlabel('k modes')
ax2.set_ylabel('$E_{mag}$')
eps = 0.5

#fig3 = plt.figure(figsize=(12,8), dpi=100)
ax3.set_xscale('log')
ax3.set_yscale('log')
ax3.set_ylim(1.0e-12, 1.0e-3)
ax3.set_xlim(1,60)
ax3.set_xlabel('k modes')
ax3.set_ylabel('$E_{mag}$')

# dicts for data storage
tdict = {}
pwdict = {}
kmod = {}

for dirs in dirlist:
    ddir = os.path.join(dirs, 'data')
    #read in parameters
    pars2 = pc.read_param(datadir=ddir,quiet=True, param2=True)
    #read in the data
    tdict[str(dirs)], pwdict[str(dirs)] = pc.read_power('power_mag.dat', datadir=ddir)
    #look for the plotting index
    pind = np.where( (tdict[str(dirs)] < 50+eps) & (tdict[str(dirs)] > 50-eps) )[0][0]
    #plot the data
    ax2.plot(pwdict[str(dirs)][pind,:], label=dirs.split('_')[1])
    ax3.plot(pwdict[str(dirs)][0,:], label=dirs.split('_')[1])
    kmod[str(dirs)] = [i/int(dirs.split('_')[1][1:]) for i in range(1,61)]
    
ax2.plot(range(1,20), [1e-12*i**2 for i in range(1,20)], color='black') 
ax2.text(5,4.5e-11, '$\propto\ k^2$', color='black')
ax3.plot(range(1,9), [5e-11*i**3 for i in range(1,9)], color='black') 
ax3.text(4,1e-8, '$\propto\ k^3$', color='black')
ax2.text(1.1, 1.0e-4, 't={0:.0f}'.format(tdict[dirlist[-1]][pind]))
ax3.text(1.1, 1.0e-4, 't={0:.0f}'.format(tdict[dirlist[-1]][0]))
ax2.legend(loc=3, prop={'size':16})
ax3.legend(loc=3, prop={'size':16})
```

Compensating the k modes in the spectra
-------------------------------------
In the paper all simulations were not plotted with the k modes on the x-Axis but only the ratio $k/k_0$ <br>
So to be able to better compare the simulations every k mode was divided by the appropriate $k_{forcing}$ number.<br>
Now the plots align very much. Although there is still is a difference in which they height of the overall spectrum. <br>
As all our simulations were only run with $N_x = 120$ and our k modes then run from $k=1,...,N_x/2$,<br>
the peak of the $k_{av}=60$ is not visible any longer. This is what was used in the paper, but with higher resolution. 

```python
fig4 = plt.figure(figsize=(12,8), dpi=100)
axx = fig4.add_subplot(111)
axx.set_xscale('log')
axx.set_yscale('log')
axx.set_xlim(.1,2)
axx.set_ylim(1.0e-11,1.0e-3)
axx.set_xlabel('$k/k_0$')
axx.set_ylabel('$E_{kin}$')
kmodes20 = [i/20. for i in range(1,61)]
kmodes10 = [i/10.15 for i in range(1,61)]
kmodes30 = [i/30. for i in range(1,61)]
kmodes60 = [i/60. for i in range(1,61)]
axx.plot(kmodes10, pwdict['rerun_k10'][pind,:], label='$k_{av}=10$')
axx.plot(kmodes20, pwdict['rerun_k20'][pind,:], label='$k_{av}=20$')
axx.plot(kmodes30, pwdict['rerun_k30'][pind,:], label='$k_{av}=30$')
axx.plot(kmodes60, pwdict['rerun_k60'][pind,:], label='$k_{av}=60$', color='purple')
axx.legend(loc=0, prop={'size':18})
axx.plot(1,1e-9, 'ko', marker=r'$\uparrow$', markersize=28)
axx.text(0.9,2e-10, 'forcing')
axx.set_title('Compensating the k modes by the factor $k_{av}$')
```
