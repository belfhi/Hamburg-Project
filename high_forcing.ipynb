High Forcing Simulation
-------------------------
To experiment with the parameters of the simulation, another run was created but now with a higher forcing parameter _force_ in the pencil-code <br>
While the other simulations ran with $f_{forcing}=1\times 10^{-3}$, this run had $f_{forcing}=7\times 10^{-2}$ <br>
This resulted in much higher $u_{rms}$ and also maximum velocities of the gas $u_{max}$.

```python
import matplotlib
%matplotlib inline
```

```python
matplotlib.rcParams.update({'font.size': 18, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
#import some modules
import pencil as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
#read in the data
t1, pkin = pc.read_power('power_kin.dat')
t2, pmag = pc.read_power('power_mag.dat')
# and the timeseries
tser = pc.read_ts(quiet=True, plot_data=False)

#read in Parameter
pars = pc.read_param(quiet=True)
par2 = pc.read_param(quiet=True, param2=True)
dim = pc.read_dim()
```

```python
fig2 = plt.figure(figsize=(12,10))
fig2.suptitle('Timeseries Plots for Velocity and magnetic Field', fontsize=26)

ts1 = fig2.add_subplot(211)
ts1.plot(tser.t, tser.urms, label='$u_{rms}$')
ts1.plot(tser.t,tser.umax, label='$u_{max}$')
ts1.set_ylabel('$u\ [c_s]$', fontdict={'size':20})
ts1.set_xlabel('t [code units]', fontdict={'size':20})
ts1.legend(loc=4)

ts2 = fig2.add_subplot(212)
ts2.plot(tser.t, tser.brms, label='$B_{rms}$')
ts2.set_ylabel('$B_{rms}$', fontdict={'size':20})
ts2.set_xlabel('t [code units]', fontdict={'size':20})
ts2.legend(loc=4)

```

Plotting the Spectra
---------------------
Now we want to plot the magnetic & kinetic spectra and compare them to the others we got from the simulations with lower forcing $f_{forcing}$

```python
#create figure for spectra
fig = plt.figure(figsize=(15,9))
fig.suptitle('Initial and relaxed Energy Spectrum', fontsize=26)

#create subplot and set up plotting area
ax1 = fig.add_subplot(221)
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_ylim(1.0e-8, 5.0e-2)
ax1.set_xlim(1,140)
ax1.set_xlabel('k modes')
ax1.set_ylabel('$E_{kin}\ &\ E_{mag}$')

#kmodes = [a/10. for a in range(144)]
kmodes = range(dim.nxgrid/2)

ax1.plot(kmodes,pkin[0,:], label='$E_{kin}$')
ax1.plot(kmodes,pmag[0,:], label='$E_{mag}$')
ax1.plot(range(1,10), [5.0e-8*i**4 for i in range(1,10)])
ax1.plot(range(1,10), [1.0e-5*i**2 for i in range(1,10)])
ax1.text(2,1e-4,'$\propto\ k^{2}$', color='cyan')
ax1.text(3,1e-6,'$\propto\ k^{4}$', color='red')
ax1.text(1.1,1.5e-2,'t= {0:.0f}'.format(t1[0]))
ax1.legend(loc=1, prop={'size':14})

ax2 = fig.add_subplot(222)
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_ylim(1.0e-8, 5.0e-2)
ax2.set_xlim(1,140)
ax2.set_xlabel('k modes')
ax2.plot(kmodes,pkin[-1,:], label='$E_{kin}$')
ax2.plot(kmodes,pmag[-1,:], label='$E_{mag}$')
ax2.plot(range(1,10), [2.0e-5*i**2 for i in range(1,10)])
ax2.text(3,5e-5,'$\propto\ k^{2}$', color='red')
ax2.text(1.1,1.5e-2,'t= {0:.0f}'.format(t1[-1]))
ax2.legend(loc=1, prop={'size':14})
```

Results:
-------
The Spectra show a remarkable $E_{kin}\propto\ k^2$ and $E_{mag}\propto\ k^4$ behaviour for very small times. <br>
This is lost at later times and the only thing that is left is the $E{mag / kin}\propto k^2$ behaviour.
