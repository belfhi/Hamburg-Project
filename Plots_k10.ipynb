Decay of Turbulence
-------------------

In this simulation the decay was investigated. 
A simulation with $N_x=120$ with driven turbulence and a forcing wavevector $k_{av}=10$ <br>
After a simulation time of $t=1000$ the forcing is turned off and the turbulence is left to decay freely.

```python
pars = pc.read_param(quiet=True)
pars2 = pc.read_param(quiet=True, param2=True)
dim = pc.read_dim()
print 'Forcing is stopped at t={0:.0f}'.format(pars2.tforce_stop)
print 'Simulation has a grid with N={0} grid points per dimension'.format(dim.nxgrid)
```

```python
t1, powerhel_kin = pc.read_power('powerhel_kin.dat')
t2, powerhel_mag = pc.read_power('powerhel_mag.dat')
t3, powermag = pc.read_power('power_mag.dat')
t4, powerkin = pc.read_power('power_kin.dat')

print powerkin.shape, t4[-1], np.where((t4 > 999) & (t4 < 1001))
print powerkin[0,:4]
```

```python
kmodes = range(dim.nxgrid/2)

fig1 = plt.figure(figsize=(12,8), dpi=100)
ax1 = fig1.add_subplot(111)
#ax1.plot(kmodes, powerkin[0,:], label='t={0:.0f}'.format(t4[0]))
for i in range(200, 300, 10): # powerkin.shape[0]+1,10):
    ax1.plot(kmodes, powerkin[i,:], label='t={0:.0f}'.format(t4[i]))
    
ax1.legend(loc=1, prop={'size':16})
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_title('Decaying Kinetic Energy Spectra')
ax1.set_ylim(1.0e-12, 1.0e-5)
ax1.set_xlim(1,dim.nxgrid/2)
ax1.set_xlabel('k modes')
ax1.set_ylabel('$E_{kin}$')
```

```python
fig2 = plt.figure(figsize=(12,8), dpi=100)
ax2 = fig2.add_subplot(111)
#ax1.plot(kmodes, powerkin[0,:], label='t={0:.0f}'.format(t4[0]))
for i in range(200, 250, 5): # powerkin.shape[0]+1,10):
    ax2.plot(kmodes, powermag[i,:], label='t={0:.0f}'.format(t4[i]))
    
ax2.legend(loc=1, prop={'size':16})
ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_title('Decaying Magnetic Energy')
ax2.set_xlim(2,dim.nxgrid/2)
ax2.set_ylim(1.0e-18, 1.0e-3)
ax2.set_yticks([1.e-3,1e-5,1e-7,1e-9,1e-11,1e-13,1e-15,1e-17])
ax2.set_xticks([1.,2.,5.,10.,20.,60.])
ax2.set_xticklabels(['1','2','5','10','20','60'])
ax2.set_xlabel('k modes')
ax2.set_ylabel('$E_{kin}$')
```
