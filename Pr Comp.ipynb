```python
def fmt(value, pos=None, digits=0):
    if value==0:
        return '$0$'
    else:
        a, b = '{0:.{1}e}'.format(value, digits).split('e')
        b = int(b)
        s = '{} \\times 10^{{{}}}'.format(a, b)
        return s
```

```python
def plot_spectra(dd):
    ddir = dd+'/data'
    pt = None
    fig = plt.figure(figsize=(10,7))
    ax = fig.add_subplot(111)
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_xlim(1,144)
    ax.set_ylim(1e-9,1e-3)
    ax.set_xlabel('k mode')
    ax.set_ylabel('$E_{mag}(k)$')
    t, powb = pc.read_power(datadir=ddir, filename='power_mag.dat')
    par2 = pc.read_param(datadir=ddir, quiet=True, param2=True)
    prM = par2.nu_hyper3/par2.eta_hyper3
    for tt, pb in enumerate(powb):
        if int(t[tt]) in [0,1,2,5,10,20,50,100] and int(t[tt]) != pt:
            ax.plot(pb, label='$t=%.0f$' %t[tt], linewidth=2)
            pt = int(t[tt])
    ax.legend(loc='center left', bbox_to_anchor=(1.,.5), ncol=1, prop={'size':16})
    fig.suptitle(r'$\nu_{3}=%s,\quad \eta_3=%s,\quad Pr_M=%g$' % (fmt(par2.nu_hyper3), fmt(par2.eta_hyper3),prM))
    fig.savefig('mag_spectra_k60_pr=%g.png' %prM, bbox_inches='tight')
    return fig, ax
```

```python
for dd in dirs:
    f1, ax1 = plot_spectra(dd)
```
