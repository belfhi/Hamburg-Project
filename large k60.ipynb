Large Run with Forcing at $k=60$
---------------------------------
This run was produced to get more information as the spectrum gets more detailed when the simulation resolution is more detailed

```python
import matplotlib 
%matplotlib inline
matplotlib.rcParams.update({'font.size': 22, 'font.family': 'STIXGeneral', 'mathtext.fontset': 'stix'})
```

```python
import pencil as pc
import numpy as np
import matplotlib.pyplot as plt
```

```python
t1, pkin = pc.read_power('power_kin.dat')
t2, pmag = pc.read_power('power_mag.dat')
dim = pc.read_dim()
```

```python
pars = pc.read_param(quiet=True)
par2 = pc.read_param(quiet=True, param2=True)
```

```python
#kmodes = [i/60. for i in range(1,dim.nxgrid/2+1)]
kmodes = range(dim.nxgrid/2)

fig = plt.figure(figsize=(16,12))
ax1 = fig.add_subplot(221)
ax1.plot(kmodes, pkin[0,:], label='$E_{kin}$')
ax1.plot(kmodes, pmag[0,:], label='$E_{mag}$')
ax1.plot(np.arange(1,50),[5.0e-12*i**4 for i in np.arange(1,50)], color='red')
ax1.text(15,1e-7,'$\propto\ k^{4}$', color='red')
#ax1.plot(np.arange(1,50),[1.0e-9*i**2 for i in np.arange(1,50)])

ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_xlim(1,dim.nxgrid/2)
ax1.set_ylim(1.0e-10, 1.0e-3)
ax1.set_xlabel('k modes')
ax1.legend(loc=4, prop={'size':18})
ax1.set_title('Spectra at t={0:.0f}'.format(t1[0]), fontsize=20)

ax2 = fig.add_subplot(222)
ax2.plot(kmodes, pkin[-1,:], label='$E_{kin}$')
ax2.plot(kmodes, pmag[-1,:], label='$E_{mag}$')
ax2.plot(np.arange(1,50),[1.0e-8*i**2 for i in np.arange(1,50)])
ax2.text(10,4e-7,'$\propto\ k^{2}$', color='red')
#ax2.text(1.5,1e-4,'t= {0:.0f}'.format(t1[-1]))

ax2.set_xscale('log')
ax2.set_yscale('log')
ax2.set_xlim(1,dim.nxgrid/2)
ax2.set_ylim(1.0e-10, 1.0e-3)
ax2.set_title('Spectra at t={0:.0f}'.format(t1[-1]), fontsize=20)
ax2.legend(loc=4, prop={'size':18})
fig.suptitle('Magnetic & Kinetic Spectra at different times', fontsize=24)
```
